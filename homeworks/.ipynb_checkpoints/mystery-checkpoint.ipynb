```python
i2=1-image[:,:,0]
a,b = i2.shape
#plot.imshow(i2,cmap='Greys')
points=[]
N=5000
f=0
while f < N:
    x=np.random.randint(b)
    y=np.random.randint(a)
    if i2[y,x]==1:
        points.append([x,y])
        f+=1
            
pi=np.array(points)
```

```python
plot.scatter(pi[:,0],-pi[:,1])
```

```python
N=500
u = np.random.normal(size=(2,N))
data=pi @ u / 100000
noise = np.random.normal(size=(5000,N))
data = data+(noise/2000)
data.shape
data
```

```python
v,w = np.linalg.eigh( data.T @ data)
```
